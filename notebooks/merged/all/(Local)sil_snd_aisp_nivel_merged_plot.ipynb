# 1. Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import plotly.express as px
import datetime
```

# 2. Load data

```python
merged = pd.read_csv('/home/nivelrios/documentos/Mburicao Project/data/external/sil_snd_aisp_local_nivel_merged.csv',index_col=0, parse_dates=True)
```

# 3. Plotting

## 3.1 Overview

```python
fig = make_subplots(specs=[[{"secondary_y": True}]])

fig.add_trace(
    go.Scatter(
        x=merged.index,
        y=merged["nivel"],
        mode="lines",
        line=dict(color='blue'),
        name="Nivel (m)"
    ),
    secondary_y=False
)

fig.add_trace(
    go.Bar(
        x=merged.index,
        y=merged["sil"],
        name="SIL (mm)",
        marker=dict(
            color="orange",
            opacity=1,
            line=dict(
                color="orange",
                width=1
            )
        )
    ),
    secondary_y=True
)

fig.add_trace(
    go.Bar(
        x=merged.index,
        y=merged["snd"],
        name="SND (mm)",
        marker=dict(
            color="darkgreen",
            opacity=1,
            line=dict(
                color="darkgreen",
                width=1
            )
        )
    ),
    secondary_y=True
)

fig.add_trace(
    go.Bar(
        x=merged.index,
        y=merged["aisp"],
        name="AISP (mm)",
        marker=dict(
            color="purple",
            opacity=1,
            line=dict(
                color="purple",
                width=1
            )
        )
    ),
    secondary_y=True
)

fig.update_layout(
   
    title_font=dict(size=24, family="Arial", weight="bold", color="black"),  # Título en negro y más grande
    xaxis_title="Fecha",
    xaxis_title_font=dict(size=18, family="Arial", weight="bold", color="black"),  # Título del eje x en negro y más grande
    xaxis=dict(
        tickfont=dict(size=14, family="Arial", color="black",weight="bold")  # Fechas en el eje X en negro y tamaño más grande
    ),
    barmode="group", 
    legend=dict(
        x=1.02,
        y=1,
        bgcolor='rgba(255,255,255,0)',
        font=dict(size=14, family="Arial", weight="bold", color="black")  # Leyenda en negro y más grande
    ),
    plot_bgcolor='white',
    height=1200,  # Ajuste del tamaño
    width=1200,  # Ajuste del tamaño
    paper_bgcolor='white',
)

fig.update_yaxes(
    title_text="Nivel (m)",
    title_font=dict(size=18, family="Arial", weight="bold", color="black"),  # Escala en negro y más grande
    tickfont=dict(color="black",weight="bold",size=14),  # Números en negro
    secondary_y=False
)

fig.update_yaxes(
    title_text="Estaciones (mm)",
    title_font=dict(size=18, family="Arial", weight="bold", color="black"),  # Escala en negro y más grande
    tickfont=dict(color="black",weight="bold",size=14),  # Números en negro
    secondary_y=True
)

fig.show()
```

## 3.2 Zoom selection

```python
def pedir_fecha(mensaje):
    print(mensaje)
    year = int(input("Año: "))
    month = int(input("Mes: "))
    day = int(input("Día: "))
    hour = int(input("Hora: "))
    minute = int(input("Minuto: "))
    second = int(input("Segundo: "))
    return datetime.datetime(year, month, day, hour, minute, second)
```

```python
fecha_inicio = pedir_fecha("Ingrese la fecha de inicio:")
fecha_fin = pedir_fecha("Ingrese la fecha de fin:")

merged_filtered = merged.loc[(merged.index >= fecha_inicio) & (merged.index <= fecha_fin)]
fig = make_subplots(specs=[[{"secondary_y": True}]])

fig.add_trace(
    go.Scatter(
        x=merged_filtered.index,
        y=merged_filtered["nivel"],
        mode="lines",
        line=dict(color='blue'),
        name="Nivel (m)"
    ),
    secondary_y=False
)

fig.add_trace(
    go.Bar(
        x=merged_filtered.index,
        y=merged_filtered["sil"],
        name="SIL (mm)",
        marker=dict(
            color="orange",
            opacity=1,
            line=dict(
                color="orange",
                width=1
            )
        )
    ),
    secondary_y=True
)

fig.add_trace(
    go.Bar(
        x=merged_filtered.index,
        y=merged_filtered["snd"],
        name="SND (mm)",
        marker=dict(
            color="darkgreen",
            opacity=1,
            line=dict(
                color="darkgreen",
                width=1
            )
        )
    ),
    secondary_y=True
)

fig.add_trace(
    go.Bar(
        x=merged_filtered.index,
        y=merged_filtered["aisp"],
        name="AISP (mm)",
        marker=dict(
            color="purple",
            opacity=1,
            line=dict(
                color="purple",
                width=2
            )
        )
    ),
    secondary_y=True
)

fig.update_layout(
    
    title_font=dict(size=24, family="Arial", weight="bold", color="black"),  # Título en negro y más grande
    xaxis_title="Fecha",
    xaxis_title_font=dict(size=20, family="Arial", weight="bold", color="black"),  # Título del eje X en negro, más grande y en negrita
    xaxis=dict(
        tickfont=dict(size=14, family="Arial", color="black",weight="bold")  # Fechas en el eje X en negro y tamaño más grande
    ),
    barmode="group",  
    legend=dict(
        x=1.02,
        y=1,
        bgcolor='rgba(255,255,255,0)',
        font=dict(size=16, family="Arial", weight="bold", color="black")  # Leyenda en negro, más grande y en negrita
    ),
    plot_bgcolor='white',
    paper_bgcolor='white',
    height=1200,  # Ajuste del tamaño
    width=1200,  # Ajuste del tamaño
)

fig.update_yaxes(
    title_text="Nivel (m)",
    title_font=dict(size=18, family="Arial", weight="bold", color="black"),  # Escala en negro y más grande
    tickfont=dict(color="black",weight="bold",size=14),  # Números en negro
    secondary_y=False
)

fig.update_yaxes(
    title_text="Estaciones (mm)",
    title_font=dict(size=18, family="Arial", weight="bold", color="black"),  # Escala en negro y más grande
    tickfont=dict(color="black",weight="bold",size=14),  # Números en negro
    secondary_y=True
)

fig.show()

```
