```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.graph_objects as go
```

```python
delay = np.arange(0, 90, 10)  # Delay en pasos de 10

sil_correlation = [0.254473, 0.348059, 0.464133, 0.565674, 0.628253, 0.638441, 0.599663, 0.530812, 0.459639]
snd_correlation = [0.270920, 0.307067, 0.356630, 0.437202, 0.517433, 0.520488, 0.498943, 0.474512, 0.468385]
aisp_correlation = [0.138523, 0.140335, 0.144304, 0.150528, 0.156753, 0.165286, 0.183669, 0.216137, 0.248468]

# Crear DataFrame
df = pd.DataFrame({
    'Delay (min)': delay,
    'SIL': sil_correlation,
    'SND': snd_correlation,
    'AISP': aisp_correlation
})

# Convertir en formato de matriz para heatmap
df_pivot = df.set_index('Delay (min)').T  # Transponer para que Delay sea el eje X
    
# Graficar el heatmap con un solo color (azul)
plt.figure(figsize=(10, 3))
sns.heatmap(df_pivot, cmap='Reds', annot=True, linewidths=0.5, fmt=".4f",
            cbar_kws={'label': 'Cross-Correlation'}, vmin=0, vmax=1)

plt.xlabel("Delay (min)")
plt.title("Heatmap de Cross-Correlation")
# Guardar imagen
plt.savefig("heatmap.png", dpi=300, bbox_inches="tight")  # Guardar como PNG
plt.show()
```

```python
df = pd.DataFrame({
    'Delay (min)': delay,
    'SIL': sil_correlation,
    'SND': snd_correlation,
    'AISP': aisp_correlation
})

# Filtrar solo la columna SIL
df_sil = df[['Delay (min)', 'SIL']]

# Convertir en formato de matriz para heatmap (transponer)
df_sil_pivot = df_sil.set_index('Delay (min)').T

# Ajustar el tamaño de la figura para que los cuadros sean más grandes
fig, ax = plt.subplots(figsize=(8, 3))  # Ajustamos la figura para reducir la altura

# Graficar el heatmap con solo la columna SIL
sns.heatmap(df_sil_pivot, cmap='Reds', annot=True, linewidths=0.5, fmt=".3f",
            cbar_kws={'label': '', 'shrink': 0.4, 'ticks': [0, 1]}, vmin=0, vmax=1, square=True, ax=ax)

# Poner en negrita los títulos y las etiquetas de los ejes
plt.xlabel("Delay (min)", fontweight='bold')

# Poner en negrita la palabra "SIL" en el eje Y
ax.set_yticklabels(['SIL'], fontweight='bold')

# Cambiar el tamaño de la fuente de la barra de color
cbar = ax.collections[0].colorbar
cbar.ax.tick_params(labelsize=12)  # Ajustar tamaño de las marcas en la barra de color

# Ajustar los márgenes de la gráfica para mejorar el espaciado
plt.subplots_adjust(left=0.1, right=0.9, top=0.8, bottom=0.3)

# Guardar imagen
plt.savefig("heatmap_sil_adjusted_bold_labels.png", dpi=300, bbox_inches="tight")
plt.show()
```
